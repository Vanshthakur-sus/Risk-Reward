```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
BEML = yf.download("BEML.NS")
```

```python
tink=BEML["Close"]
tink

```

```python
tink["SMA50"]=tink.rolling(window=50, min_periods=50).mean()
tink
```

```python
tink.plot(figsize=(12,8), fontsize=(15))
plt.legend(loc="upper left", fontsize=15)
plt.show()
```
