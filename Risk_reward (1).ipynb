```python
#PLOTTING NORMALIZED STOCK DATA
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt


def normalize_stock_data(ticker, start_date, end_date):
    # Fetch stock data
    stock_data = yf.download(ticker, start=start_date, end=end_date)

    # Extract Close Prices
    close_prices = stock_data['Close']

    # Min-Max Normalization: (X - min) / (max - min)
    normalized_prices = (close_prices - close_prices.min()) / (close_prices.max() - close_prices.min())

    # Plot the normalized stock prices
    plt.figure(figsize=(10,5))
    plt.plot(normalized_prices, label=f'Normalized {ticker}')
    plt.title(f'Normalized Stock Prices of {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Normalized Price')
    plt.legend()
    plt.show()

    return normalized_prices

# Example Usage
ticker =  ['RELIGARE.NS','SUNPHARMA.NS'] # Change to any stock symbol
start_date = "2025-03-31"
end_date = "2025-04-04"

normalized_data = normalize_stock_data(ticker, start_date, end_date)
print(normalized_data.head())  # Display first few normalized values

```
